```python
#<aside>
ðŸ’¡ 1. **Merge k Sorted Lists**

You are given an array of `k` linked-lists `lists`, each linked-list is sorted in ascending order.

*Merge all the linked-lists into one sorted linked-list and return it.*

**Example 1:**

Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted list:
1->1->2->3->4->4->5->6

Example 2:                      
    
Input: lists = []
Output: []
    
Example 3:                       
    
Input: lists = [[]]
Output: []
    
**Constraints:**

- `k == lists.length`
- `0 <= k <= 10000`
- `0 <= lists[i].length <= 500`
- `-10000 <= lists[i][j] <= 10000`
- `lists[i]` is sorted in **ascending order**.
- The sum of `lists[i].length` will not exceed `10000`.

</aside> 

"""To merge k sorted linked-lists into one sorted linked-list, you can use a Min Heap (Priority Queue) approach."""

# Here's a Python function that implements this algorithm:

import heapq

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def merge_k_sorted_lists(lists):
    heap = []
    for l in lists:
        while l:
            heapq.heappush(heap, l.val)
            l = l.next
    
    dummy_head = ListNode()
    current = dummy_head
    while heap:
        current.next = ListNode(heapq.heappop(heap))
        current = current.next
    
    return dummy_head.next
# Define linked lists
list1 = ListNode(1, ListNode(4, ListNode(5)))
list2 = ListNode(1, ListNode(3, ListNode(4)))
list3 = ListNode(2, ListNode(6))

lists = [list1, list2, list3]

result = merge_k_sorted_lists(lists)
# Output: 1 -> 1 -> 2 -> 3 -> 4 -> 4 -> 5 -> 6
lists = []
result = merge_k_sorted_lists(lists)
# Output: None (An empty linked list)
lists = [[]]
result = merge_k_sorted_lists(lists)
# Output: None (An empty linked list)
def print_linked_list(head):
    values = []
    while head:
        values.append(head.val)
        head = head.next
    print(values)

"""We'll need to create a ListNode class to represent the elements of the linked lists. The function merge_k_sorted_lists
   uses a min heap to efficiently merge the linked lists' values while keeping the sorted order. It iterates over each
   linked list in the input, pushing its elements' values into the min heap. Then it creates a new sorted linked list
   by popping the elements from the heap and linking them together."""    
```

```python
#<aside>
ðŸ’¡ 2. **Count of Smaller Numbers After Self**

Given an integer array `nums`, return *an integer array* `counts` *where* `counts[i]` *is the number of smaller
elements to the right of* `nums[i]`.

**Example 1:**

Input: nums = [5,2,6,1]
Output: [2,1,1,0]
Explanation:
To the right of 5 there are2 smaller elements (2 and 1).
To the right of 2 there is only1 smaller element (1).
To the right of 6 there is1 smaller element (1).
To the right of 1 there is0 smaller element.

Example 2:                       
    
Input: nums = [-1]
Output: [0]
    
Example 3:                     
    
Input: nums = [-1,-1]
Output: [0,0]
    
**Constraints:**

- `1 <= nums.length <= 100000`
- `-10000 <= nums[i] <= 10000`

</aside> 

"""To solve the "Count of Smaller Numbers After Self" problem, you can use the "Merge Sort" approach with a
   modified merge step. This approach efficiently counts the number of smaller elements to the right of each
   element in the array."""

#Here's a Python function to achieve this:

def count_smaller(nums):
    def merge_sort(enum):
        half = len(enum) // 2
        if half:
            left, right = merge_sort(enum[:half]), merge_sort(enum[half:])
            for i in range(len(enum) - 1, -1, -1):
                if not right or (left and left[-1][1] > right[-1][1]):
                    smaller_count[left[-1][0]] += len(right)
                    enum[i] = left.pop()
                else:
                    enum[i] = right.pop()
        return enum

    smaller_count = [0] * len(nums)
    merge_sort(list(enumerate(nums)))
    return smaller_count
# Test case 1
nums = [5, 2, 6, 1]
result = count_smaller(nums)
# Output: [2, 1, 1, 0]

# Test case 2
nums = [-1]
result = count_smaller(nums)
# Output: [0]

# Test case 3
nums = [-1, -1]
result = count_smaller(nums)
# Output: [0, 0]

"""The count_smaller function uses the merge_sort helper function to sort the array while simultaneously updating
   the smaller_count array. During the merge step, it counts the number of elements in the right subarray that are 
   smaller than the current element in the left subarray and updates the smaller_count accordingly."""
```

```python
#<aside>
ðŸ’¡ 3. **Sort an Array**

Given an array of integers `nums`, sort the array in ascending order and return it.

You must solve the problem **without using any built-in** functions in `O(nlog(n))` time complexity and with the
smallest space complexity possible.

**Example 1:
    
Input: nums = [5,2,3,1]
Output: [1,2,3,5]
Explanation: After sorting the array, the positions of some numbers are not changed (for example, 2 and 3), while 
the positions of other numbers are changed (for example, 1 and 5).

Example 2:                        
    
Input: nums = [5,1,1,2,0,0]
Output: [0,0,1,1,2,5]
Explanation: Note that the values of nums are not necessairly unique.
**Constraints:**

- `1 <= nums.length <= 5 * 10000`
- `-5 * 104 <= nums[i] <= 5 * 10000`
</aside>

"""To sort an array of integers nums in ascending order without using any built-in functions in O(n log n) time 
   complexity and with minimal space complexity, you can implement the Merge Sort algorithm. Merge Sort is a stable,
   comparison-based sorting algorithm that divides the input array into two halves, sorts them independently, and
   then merges them back together."""

#Here's a Python function to implement Merge Sort:

def sort_array(nums):
    def merge_sort(arr):
        if len(arr) <= 1:
            return arr
        
        mid = len(arr) // 2
        left = merge_sort(arr[:mid])
        right = merge_sort(arr[mid:])
        
        return merge(left, right)
    
    def merge(left, right):
        merged = []
        i, j = 0, 0
        
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                merged.append(left[i])
                i += 1
            else:
                merged.append(right[j])
                j += 1
        
        merged.extend(left[i:])
        merged.extend(right[j:])
        return merged
    
    return merge_sort(nums)
nums = [5, 2, 3, 1]
result = sort_array(nums)
# Output: [1, 2, 3, 5]
nums = [5, 1, 1, 2, 0, 0]
result = sort_array(nums)
# Output: [0, 0, 1, 1, 2, 5]

"""The function should return the sorted array for both examples.

   Merge Sort has a time complexity of O(n log n) and a space complexity of O(n), which satisfies the requirements
   of the problem. It is a stable sorting algorithm, which means the relative order of equal elements is preserved.
   This implementation is efficient for the given constraints."""
```

```python
#<aside>
ðŸ’¡ 4. **Move all zeroes to end of array**

Given an array of random numbers, Push all the zeroâ€™s of a given array to the end of the array. For example,
if the given arrays is {1, 9, 8, 4, 0, 0, 2, 7, 0, 6, 0}, it should be changed to {1, 9, 8, 4, 2, 7, 6, 0, 0, 0, 0}. 
The order of all other elements should be same. Expected time complexity is O(n) and extra space is O(1).

**Example:**

Input :  arr[] = {1, 2, 0, 4, 3, 0, 5, 0};
Output : arr[] = {1, 2, 4, 3, 5, 0, 0, 0};

Input : arr[]  = {1, 2, 0, 0, 0, 3, 6};
Output : arr[] = {1, 2, 3, 6, 0, 0, 0};

</aside>     

"""To move all zeroes to the end of an array with an expected time complexity of O(n) and extra space of O(1),
   you can use a two-pointer approach. This approach efficiently moves non-zero elements to the front of the
   array while keeping track of the number of zeroes encountered. Once all non-zero elements are moved to the 
   front, you can fill the remaining elements with zeroes."""

#Here's a Python function to implement this:

def move_zeroes_to_end(arr):
    n = len(arr)
    zero_count = 0

    # Move non-zero elements to the front
    for i in range(n):
        if arr[i] != 0:
            arr[zero_count], arr[i] = arr[i], arr[zero_count]
            zero_count += 1

    # Fill the remaining elements with zeroes
    for i in range(zero_count, n):
        arr[i] = 0

    return arr
arr = [1, 2, 0, 4, 3, 0, 5, 0]
result = move_zeroes_to_end(arr)
# Output: [1, 2, 4, 3, 5, 0, 0, 0]
arr = [1, 2, 0, 0, 0, 3, 6]
result = move_zeroes_to_end(arr)
# Output: [1, 2, 3, 6, 0, 0, 0]

"""The function should return the arrays with all zeroes moved to the end while preserving the order of other elements.

   This algorithm has a time complexity of O(n) since it requires a single pass through the array, and it uses constant 
   extra space (O(1)), fulfilling the given requirements."""
```

```python
#<aside>
ðŸ’¡ 5. **Rearrange array in alternating positive & negative items with O(1) extra space**

Given an **array of positive** and **negative numbers**, arrange them in an **alternate** fashion such that 
every positive number is followed by a negative and vice-versa maintaining the **order of appearance**. 
The number of positive and negative numbers need not be equal. If there are more positive numbers they
appear at the end of the array. If there are more negative numbers, they too appear at the end of the array.

**Examples:**

Input:  arr[] = {1, 2, 3, -4, -1, 4}
Output: arr[] = {-4, 1, -1, 2, 3, 4}

Input:  arr[] = {-5, -2, 5, 2, 4, 7, 1, 8, 0, -8}
Output: arr[] = {-5, 5, -2, 2, -8, 4, 7, 1, 8, 0}

</aside> 

"""To rearrange an array of positive and negative numbers in an alternate fashion while maintaining the order of
   appearance, you can use a two-pointer approach. The idea is to find the first positive number and the first 
   negative number in the array and then swap them. Repeat this process until all elements are rearranged."""

#Here's a Python function to implement this:

def rearrange_alternating_positive_negative(arr):
    n = len(arr)
    i, j = 0, 0

    # Find the first positive number
    while i < n and arr[i] >= 0:
        i += 1

    # Find the first negative number
    while j < n and arr[j] < 0:
        j += 1

    while i < n and j < n:
        # Swap the positive and negative numbers
        arr[i], arr[j] = arr[j], arr[i]

        # Find the next positive number
        while i < n and arr[i] >= 0:
            i += 1

        # Find the next negative number
        while j < n and arr[j] < 0:
            j += 1

    return arr
arr = [1, 2, 3, -4, -1, 4]
result = rearrange_alternating_positive_negative(arr)
# Output: [-4, 1, -1, 2, 3, 4]
arr = [-5, -2, 5, 2, 4, 7, 1, 8, 0, -8]
result = rearrange_alternating_positive_negative(arr)
# Output: [-5, 5, -2, 2, -8, 4, 7, 1, 8, 0]


"""The function should return the arrays with positive and negative numbers arranged in an alternating fashion 
   while preserving the order of appearance.

   This algorithm has a time complexity of O(n) since it requires a single pass through the array, and it uses
   constant extra space (O(1)), fulfilling the given requirements."""
```

```python
#<aside>
ðŸ’¡ **6. Merge two sorted arrays**

Given two sorted arrays, the task is to merge them in a sorted manner.

**Examples:**

> Input: arr1[] = { 1, 3, 4, 5}, arr2[] = {2, 4, 6, 8} 
Output: arr3[] = {1, 2, 3, 4, 4, 5, 6, 8}

Input: arr1[] = { 5, 8, 9}, arr2[] = {4, 7, 8}
Output: arr3[] = {4, 5, 7, 8, 8, 9}
> 
</aside>

"""To merge two sorted arrays into a single sorted array, you can use a two-pointer approach. The idea is to compare 
   elements from both arrays and insert them into a new array in a sorted manner."""

#Here's a Python function to implement this:

def merge_sorted_arrays(arr1, arr2):
    n1, n2 = len(arr1), len(arr2)
    merged = []

    i, j = 0, 0
    while i < n1 and j < n2:
        if arr1[i] <= arr2[j]:
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1

    while i < n1:
        merged.append(arr1[i])
        i += 1

    while j < n2:
        merged.append(arr2[j])
        j += 1

    return merged
arr1 = [1, 3, 4, 5]
arr2 = [2, 4, 6, 8]
result = merge_sorted_arrays(arr1, arr2)
# Output: [1, 2, 3, 4, 4, 5, 6, 8]
arr1 = [5, 8, 9]
arr2 = [4, 7, 8]
result = merge_sorted_arrays(arr1, arr2)
# Output: [4, 5, 7, 8, 8, 9]

"""The function should return a new array that contains elements from both arrays, merged in a sorted manner.

   The time complexity of this approach is O(n1 + n2), where n1 and n2 are the lengths of the two arrays, and the 
   extra space required is proportional to the combined lengths of the input arrays, fulfilling the given requirements."""
```

```python
#<aside>
ðŸ’¡ 7. **Intersection of Two Arrays**

Given two integer arrays `nums1` and `nums2`, return *an array of their intersection*. Each element in the result
must be **unique** and you may return the result in **any order**.

**Example 1:**
    
Input: nums1 = [1,2,2,1], nums2 = [2,2]
Output: [2]
    
Example 2:               
    
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]
Output: [9,4]
Explanation: [4,9] is also accepted.
    
**Constraints:**

</aside>  

"""To find the intersection of two integer arrays nums1 and nums2, you can use Python's built-in set data structure. 
   Sets automatically handle duplicates and provide efficient operations for finding intersections."""

#Here's a Python function to implement this:

- `1 <= nums1.length, nums2.length <= 1000`
- `0 <= nums1[i], nums2[i] <= 1000

def intersection(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    return list(set1.intersection(set2))
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]
result = intersection(nums1, nums2)
# Output: [2]
nums1 = [4, 9, 5]
nums2 = [9, 4, 9, 8, 4]
result = intersection(nums1, nums2)
# Output: [9, 4]


"""The function should return an array that contains the unique elements present in both nums1 and nums2. 
   The order of elements in the result may vary, but each element must be unique.

   The time complexity of this approach is O(n + m), where n and m are the lengths of nums1 and nums2, 
   respectively, because converting a list to a set takes O(n) time, and finding the intersection of two 
   sets also takes O(min(n, m)) time. The space complexity is O(min(n, m)) due to the storage of sets."""
```

```python
#<aside>
ðŸ’¡ 8. **Intersection of Two Arrays II**

Given two integer arrays `nums1` and `nums2`, return *an array of their intersection*. Each element in the result 
must appear as many times as it shows in both arrays and you may return the result in **any order**.

**Example 1:**

Input: nums1 = [1,2,2,1], nums2 = [2,2]
Output: [2,2]
    
Example 2:       
    
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]
Output: [4,9]
Explanation: [9,4] is also accepted.
    
**Constraints:**

- `1 <= nums1.length, nums2.length <= 1000`
- `0 <= nums1[i], nums2[i] <= 1000`

</aside>

"""To find the intersection of two integer arrays nums1 and nums2, considering the elements should appear as many
   times as they do in both arrays, you can use Python's built-in Counter data structure from the collections module. 
   The Counter allows counting the occurrences of elements in a list."""

#Here's a Python function to implement this:

from collections import Counter

def intersection(nums1, nums2):
    count1 = Counter(nums1)
    count2 = Counter(nums2)
    intersection_count = count1 & count2
    result = []
    for num, count in intersection_count.items():
        result.extend([num] * count)
    return result
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]
result = intersection(nums1, nums2)
# Output: [2, 2]
nums1 = [4, 9, 5]
nums2 = [9, 4, 9, 8, 4]
result = intersection(nums1, nums2)
# Output: [4, 9]


"""The function should return an array that contains the elements that appear as many times as they do in both 
   nums1 and nums2. The order of elements in the result may vary, but each element must appear as many times as 
   it does in both arrays.

   The time complexity of this approach is O(n + m), where n and m are the lengths of nums1 and nums2, respectively,
   because constructing the counters takes O(n + m) time, and finding the intersection of two counters also takes
   O(min(n, m)) time. The space complexity is O(n + m) due to the storage of counters."""
```
